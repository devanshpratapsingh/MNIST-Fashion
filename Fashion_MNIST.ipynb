```python
# Importing required libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
training_df = pd.read_csv('fashion-mnist_train.csv')
testing_df = pd.read_csv('fashion-mnist_test.csv')
```

```python
trainData = np.array(training_df, dtype='float32')
testData = np.array(testing_df, dtype='float32')
```

```python
# to view multiple images in a grid format

Grid_W = 10 
Grid_H = 10

#creating subplot to display multiple images
fig, axes = plt.subplots(nrows=Grid_H, ncols=Grid_W, figsize = (18,18))
axes = axes.ravel()

#loop through the training dataset to display random 100 images
for i in np.arange(0, Grid_W*Grid_H):
    index = np.random.randint(0, len(trainData))
    axes[i].imshow(trainData[index, 1:].reshape(28, 28))
    axes[i].set_title(trainData[index, 0])
    axes[i].axis('off')   
```

```python
# spliting the data into initial X_train and y_train data
# Normalizing each pixel value by dividing the same with 255
# resulting in converting all pixel values between range 0 to 1

X_train_init = trainData[:, 1:] / 255 
y_train_init = trainData[:, 0]

X_test = testData[:, 1:] / 255 
y_test = testData[:, 0]
```

```python
#splitting the training data further to create training and validation data

from sklearn.model_selection import train_test_split
X_train, X_valid, y_train, y_valid = train_test_split(X_train_init, y_train_init, 
                                                      test_size=0.2, random_state = 0)
```

```python
X_train.shape
```

```python
X_valid.shape
```

```python
# reshaping the array silimar to image data
X_train = X_train.reshape(X_train.shape[0], 28, 28, 1)
X_test = X_test.reshape(X_test.shape[0], 28, 28, 1)
X_valid = X_valid.reshape(X_valid.shape[0], 28, 28, 1)
```

```python
# new shape of the training, testing and validation data

print(X_train.shape)
print(X_test.shape)
print(X_valid.shape)
```

```python
# building our neural network

import keras
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, GaussianNoise
from tensorflow.keras.optimizers import Adam
from keras.layers.normalization import BatchNormalization
```

```python
model=Sequential()
model.add(Conv2D(32,kernel_size=(3,3),activation='relu',padding='same',input_shape=(28,28,1)))
model.add(Conv2D(32,kernel_size=(3,3),activation='relu',padding='same'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.15))
model.add(BatchNormalization())
model.add(Conv2D(64,kernel_size=(3,3),activation='relu',padding='same'))
model.add(Conv2D(64,kernel_size=(3,3),activation='relu',padding='same'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.15))
model.add(BatchNormalization())
model.add(GaussianNoise(0.25))
model.add(Flatten())
model.add(Dense(64,activation='relu'))
model.add(Dropout(0.25))
model.add(Dense(64,activation='relu'))
model.add(Dropout(0.25))
model.add(BatchNormalization())
model.add(GaussianNoise(0.25))
model.add(Dense(10,activation='softmax'))

model.summary()
```

```python
model.compile(optimizer='Adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
history = model.fit(x=X_train, 
             y=y_train, 
             batch_size=128, 
             epochs=25, 
             verbose=1, 
             validation_data=(X_valid, y_valid)
            )
```

```python
accuracy = model.evaluate(X_test, y_test)
print('Test Acuuracy : {}'.format(accuracy[1]))
```

```python
# Getting the prediction classes for test data
prediction_classes = model.predict_classes(X_test)
```

```python
# visualization of predicted classes with it's true classes

Grid_W = 6
Grid_H = 6

fig, axes = plt.subplots(Grid_H, Grid_W, figsize=(20,20))

axes = axes.ravel()

for i in np.arange(0, Grid_H*Grid_W):
    axes[i].imshow(X_test[i].reshape(28,28))
    axes[i].set_title('Predicted = {:0.1f} \n Actual = {:0.1f}'.format(prediction_classes[i], 
                                                                       y_test[i]))
    axes[i].axis('off')
```

```python
#claasification report 

from sklearn.metrics import classification_report

classes = 10
targets = ["Class {}".format(i) for i in range(classes)]

print(classification_report(y_test, prediction_classes, target_names = targets))
```
